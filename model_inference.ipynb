# Get Model

```python
import dagshub
import mlflow
from mlflow.models.signature import infer_signature
dagshub.init(repo_owner='Cimbir', repo_name='advanced-regression-techniques', mlflow=True)
```

```python
logged_model = 'models:/all_OHE_best/1'

loaded_model = mlflow.pyfunc.load_model(logged_model)
```

# Create Submission

```python
pred = loaded_model.predict(test)
pred_df = pd.DataFrame(pred)
pred_df.index += 1461  # Adjust index to start from 1
pred_df.columns = ['SalePrice']  # Rename column to 'SalePrice'
pred_df.insert(0, 'Id', pred_df.index)  # Add 'Id' column
pred_df.head()
```

```python
pred_df.to_csv('submissions/submission.csv', index=False)
print("Submission file created: submission.csv")
```
